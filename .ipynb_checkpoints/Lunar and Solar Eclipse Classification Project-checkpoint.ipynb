```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfLunar = pd.read_csv('datasets/lunar.csv')
dfLunar.head()
```

# Exploração dos Dados e Pré-Processamento (Eclipse Lunar)

```python
dfLunar.info()
```

```python
dfLunar.isnull().values.any()
```

```python
dfLunar.describe()
```

```python
dfLunar.drop(['Delta T (s)','Gamma'],axis=1,inplace=True)
dfLunar.head()
```

```python
print(dfLunar['Eclipse Type'].unique())
print(dfLunar['Quincena Solar Eclipse'].unique())
```

```python
dfLunar['Eclipse Type'] = pd.factorize(dfLunar['Eclipse Type'])[0] + 1
dfLunar['Quincena Solar Eclipse'] = pd.factorize(dfLunar['Quincena Solar Eclipse'])[0] + 1
dfLunar.head()
```

```python
dfLunar2=dfLunar
dfLunar2['Partial Eclipse Duration (m)'].str.contains('-')
```

```python
dfLunar3= dfLunar2['Partial Eclipse Duration (m)']
dfLunar3.replace('-', np.nan,inplace=True)
dfLunar3=dfLunar3.astype(float)
val = dfLunar3.mean(axis=0,skipna = True)
print(val)
```

```python
dfLunar2['Partial Eclipse Duration (m)'].replace(np.nan, val,inplace=True)
dfLunar2['Partial Eclipse Duration (m)']=pd.to_numeric(dfLunar2["Partial Eclipse Duration (m)"])
```

```python
dfLunar3= dfLunar2['Total Eclipse Duration (m)']
dfLunar3.replace('-', np.nan,inplace=True)
dfLunar3=dfLunar3.astype(float)
val = dfLunar3.mean(axis=0,skipna = True)
print(val)
dfLunar2['Total Eclipse Duration (m)'].replace(np.nan, val,inplace=True)
dfLunar2['Total Eclipse Duration (m)']=pd.to_numeric(dfLunar2["Total Eclipse Duration (m)"])
```

```python
dfLunar2.info()
```

```python
dfLunar2.head()
```

```python
dfLunar2['Calendar Date']=dfLunar2['Calendar Date'].str.replace('-', '')
dfLunar2.head()
```

```python
dfLunar2['Eclipse Datetime']=dfLunar2['Calendar Date']+' '+dfLunar2['Eclipse Time']
dfLunar2.head()
```

```python
dfLunar2['Eclipse Datetime']=pd.to_datetime(dfLunar2['Eclipse Datetime'],errors='coerce')
dfLunar2['Eclipse Datetime']=pd.to_numeric(dfLunar2['Eclipse Datetime'],errors='coerce')
dfLunar2.info()
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(dfLunar2.corr(),   #.corr() traz a correlação de Pearson, mostrando um relacionamento linear entre as variáveis
            annot = True,
            fmt = '.2f',
            cmap='Blues')
plt.title('Correlação entre variáveis do dataset de Eclipse Lunar')
plt.show()
```

```python
dfLunar2.drop(['Calendar Date','Eclipse Time','Latitude','Longitude','Eclipse Datetime','Partial Eclipse Duration (m)','Total Eclipse Duration (m)','Quincena Solar Eclipse'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(dfLunar2.corr(),   #.corr() traz a correlação de Pearson, mostrando um relacionamento linear entre as variáveis
            annot = True,
            fmt = '.2f',
            cmap='Blues')
plt.title('Correlação entre variáveis do dataset de Eclipse Lunar')
plt.show()
```

# Criação de Modelos de Machine Learning e Análise dos Resultados iniciais

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from pycaret.classification import *
```

```python
model = GaussianNB()
model2 = KNeighborsClassifier(n_neighbors=10)
```

```python
dfLunar2['Eclipse Type'] = dfLunar2.pop('Eclipse Type')
dfLunar2.head()
```

```python
dfLunar3 = dfLunar2.drop(['Catalog Number'],axis=1)
dfLunar3.shape
```

```python
X = dfLunar3.drop(['Eclipse Type'],axis=1)
y = dfLunar3['Eclipse Type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
X_test.head()
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

# Análise com Pycaret em busca do melhor modelo possível para o Eclipse Lunar

```python
s = setup(dfLunar3, target = 'Eclipse Type', remove_perfect_collinearity=False)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
plot_model(best, plot = 'confusion_matrix')
```

```python
predictions = predict_model(best, data=dfLunar3, raw_score=True)
predictions.head()
```

```python
print("Score do Pycaret - melhor modelo encontrado:",best.score(X_test, y_test))
y_pred = best.predict(X_test)
print(classification_report(y_test, y_pred))
```
